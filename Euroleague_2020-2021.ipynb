```python
import pandas as pd
import numpy as np
```

```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args
        
    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
df1 = df['2FG'].str.split('/', expand=True).astype('int32') 
df2 = df['3FG'].str.split('/', expand=True).astype('int32')
df3 = df['FT'].str.split('/', expand=True).astype('int32')
```

```python
df4 = df1 + df2 + df3
```

```python
df.insert(13, ("Total_Shot"), df4[0])
df.insert(14, ("Total_Try"), df4[1])
df.insert(15, ("Total_Avg"), (df4[0] / df4[1]))
```
